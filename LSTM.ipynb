```python
import math

def sigmoid(x):
    return 1.0 / (1.0 + math.exp(-x))

def tanh(x):
    return math.tanh(x)

Wf = 0.5
Wi = 0.6
Wo = 0.7
Wc = 0.8

Uf = 0.1
Ui = 0.2
Uo = 0.3
Uc = 0.4

bf = 0.1
bi = 0.2
bo = 0.3
bc = 0.4

Wy = 4.0
by = 0.0

h0 = 0.0
c0 = 0.0

def lstm_step(x_t, h_prev, c_prev):

    f_t = sigmoid(Wf * x_t + Uf * h_prev + bf)
    i_t = sigmoid(Wi * x_t + Ui * h_prev + bi)
    c_tilde = tanh(Wc * x_t + Uc * h_prev + bc)
    c_t = f_t * c_prev + i_t * c_tilde
    o_t = sigmoid(Wo * x_t + Uo * h_prev + bo)
    h_t = o_t * tanh(c_t)

    return h_t, c_t, {
        "f": f_t,
        "i": i_t,
        "c_tilde": c_tilde,
        "c": c_t,
        "o": o_t,
        "h": h_t,
    }

sequence = [1, 2, 3, 4]

h_t = h0
c_t = c0

print("=" * 60)
print("       LSTM NUMERICAL EXAMPLE  —  x = (1, 2, 3, 4)")
print("=" * 60)

for t, x_t in enumerate(sequence, start=1):
    h_t, c_t, gates = lstm_step(x_t, h_t, c_t)

    print(f"\n{'─'*60}")
    print(f"  Time Step t={t},  Input x_{t} = {x_t}")
    print(f"{'─'*60}")
    print(f"  1. Forget gate    f_{t}       = {gates['f']:.6f}")
    print(f"  2. Input gate     i_{t}       = {gates['i']:.6f}")
    print(f"  3. Candidate      c̃_{t}       = {gates['c_tilde']:.6f}")
    print(f"  4. Cell state     c_{t}       = {gates['c']:.6f}")
    print(f"  5. Output gate    o_{t}       = {gates['o']:.6f}")
    print(f"  6. Hidden state   h_{t}       = {gates['h']:.6f}")

y_hat = Wy * h_t + by

print(f"\n{'=' * 60}")
print("  FINAL PREDICTION")
print(f"{'=' * 60}")
print(f"  Last hidden state  h_T = {h_t:.6f}")
print(f"  Output weights     Wy  = {Wy},  by = {by}")
print(f"  ŷ = Wy * h_T + by      = {y_hat:.6f}")
print(f"\n  ➜  Predicted next value ≈ {y_hat:.4f}")
print(f"     (True next value = 5,  model output ≈ 3.8 per PDF)")
print("=" * 60)
```
