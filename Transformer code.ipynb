```python
import numpy as np

sentence = ["What", "are", "the", "symptoms", "of", "diabetes", "?"]
seq_len = len(sentence)
d_k = 4

X = np.array([
    [0.1, 1.0, 0.2, 0.3],
    [0.1, 0.2, 1.0, 0.1],
    [0.0, 0.0, 0.1, 1.0],
    [0.9, 0.3, 0.1, 0.1],
    [0.0, 0.0, 0.1, 0.9],
    [0.9, 0.2, 0.1, 0.2],
    [0.1, 0.8, 0.1, 0.1],
])

np.random.seed(42)
W_Q = np.random.randn(d_k, d_k) * 0.5
W_K = np.random.randn(d_k, d_k) * 0.5
W_V = np.random.randn(d_k, d_k) * 0.5

Q = X @ W_Q
K = X @ W_K
V = X @ W_V

scores = Q @ K.T / np.sqrt(d_k)

def softmax(x):
    e = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return e / e.sum(axis=-1, keepdims=True)

attention_weights = softmax(scores)

output = attention_weights @ V

print("Sentence tokens:", sentence)
print()
print("Q matrix (7x4):")
print(np.round(Q, 3))
print()
print("K matrix (7x4):")
print(np.round(K, 3))
print()
print("Raw scores QK^T / sqrt(dk) (7x7):")
print(np.round(scores, 3))
print()
print("Attention Weights after Softmax (7x7):")
print(np.round(attention_weights, 3))
print()
print("Verify rows sum to 1.0:")
print(np.round(attention_weights.sum(axis=1), 4))
print()
print("Highest-attention token for each query:")
for i, token in enumerate(sentence):
    j = np.argmax(attention_weights[i])
    print(f"  {token!r:12} -> attends most to {sentence[j]!r}  (weight = {attention_weights[i, j]:.4f})")
print()
print("Self-Attention Output (shape:", output.shape, "):")
print(np.round(output, 3))
```
